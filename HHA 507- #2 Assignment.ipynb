## Loading Dataset CSV file

```python
import pandas as pd
```

```python
import re
```

```python
url = 'https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv'
df_hc= pd.read_csv(url)
```

```python
df_hc.head(10)
```

## Identifying & Handling Values

### Size & Shape Identification

```python
print("Number of rows:", df_hc.shape[0])
print("Number of columns:", df_hc.shape[1])
```

```python
df_hc.columns
```

```python
df_hc.head(10)
```

```python
df_hc.tail(10)
```

```python
df_hc.sample(10)
```

### Filling and Handling Missing Values

*Replacing "Missing" & "None" with NaN*

```python
import numpy as np
```

```python
df_hc.replace(["None","missing"], np.nan, inplace=True)
```

```python
df_hc.head(10)
```

*Replacing numerical categories with mean:*

```python
df_hc.fillna(df_hc.mean(), inplace=True)
```

*Replacing word/written caregories with the mode:*



```python
df_hc.fillna(df_hc.mode().iloc[0], inplace=True)
```

```python
df_hc.head(10)
```

## Removing any duplicate rows and columns

*Removing duplicate rows:*

```python
df_hc.drop_duplicates(inplace=True)
```

*Removing duplicate columns:*

```python
data = df_hc.T.drop_duplicates().T
```

```python
print("Number of rows:", df_hc.shape[0])
print("Number of columns:", df_hc.shape[1])
```

## Cleaning column names

*Clean Column Names of extra spaces:*

```python
df_hc.columns = df_hc.columns.str.strip()
```

*Replace spaces with underscores in column names*

```python
df_hc.columns = df_hc.columns.str.replace(' ','_')
```

```python
df_hc.head(10)
```

## Data Transformation- creating new columns based on existing names

Creating a BMI column based on the existing weight and heigh of the patients

*(Running the code below to recognize the weight and height columns as integers due to running in errors when calculating the BMI:)*

```python
df_hc['Height_(in_cm)'] = df_hc['Height_(in_cm)'].astype(int)
```

```python
df_hc['Weight_(in_kg)'] = df_hc['Weight_(in_kg)'].astype(int)
```

```python
df_hc.head(10)
```

Creating the BMI column:

```python
df_hc['BMI'] = df_hc['Weight_(in_kg)']/ ((df_hc['Height_(in_cm)']/100)**2)
```

```python
df_hc.head(10)
```

Switching the order of the BMI column and moving it next to the weight & height columns:

```python
new_column_order = ['Patient_Age','Gender','City_of_Residence','State_of_Residence','Has_Insurance','Visited_Last_Month','Payment_Method','Preferred_Doctor','Disease_Diagnosed','Medication_Prescribed','Type_of_Appointment','Average_Heart_Rate','Average_BP','Height_(in_cm)','Weight_(in_kg)','BMI','Payment_Due_($)','Last_Visit_(days_ago)','Visit_Duration_(mins)','Number_of_Tests','Prescription_Cost_($)']
df_hc = df_hc[new_column_order + [col for col in df_hc.columns if col not in new_column_order]]
```

```python
df_hc.head(10)
```

Rounding the BMI Data to the nearest hundredth:

```python
df_hc['BMI'] = df_hc['BMI'].round(3)
```

```python
df_hc.head(10)
```

###GroupBy and Aggregation Fx:

```python
print(df_hc.columns)
```

Grouped the Disease Diagnosed and found the average of the patients' age for each diasease dx:

```python
grouped = df_hc.groupby('Disease_Diagnosed')
```

```python
df_hc['Patient_Age'] = df_hc['Patient_Age'].astype(int)
```

```python
avg_age_disease = grouped['Patient_Age'].mean()
```

```python
print(avg_age_disease)
```

Insurance status average based on the age of the patient:

```python
grouped2 = df_hc.groupby('Has_Insurance')
```

```python
avg_age_insurance = grouped2['Patient_Age'].mean()
```

```python
print(avg_age_insurance)
```

### Using Pivot tables or cross-tabulations for multi-dimensional analysis

Creating a Pivot table for analyzing the patients' age diagnosed for each disease sorted by their gender:

```python
pivot_table = pd.pivot_table(df_hc,
                             values='Patient_Age',
                             index='Disease_Diagnosed',
                             columns='Gender',
                             aggfunc='mean')
```

```python
print(pivot_table)
```
